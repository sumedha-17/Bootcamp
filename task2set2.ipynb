```python
#merge two dictionaries
d1={'a':10,'b':20,'c':77}
d2={'b':70,'c':89,'d':90}
for key in d2:
    if key in d1:
        d1[key]=d1[key]+d2[key]
else:
    d1[key]=d2[key]
    print(d1)

```

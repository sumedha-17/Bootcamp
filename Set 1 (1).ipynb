```python

i=input("Enter items")
my_list=i.split()
print(my_list)
max=int(my_list[0])
min=int(my_list[0])
sum=0
for i in my_list:
    val=int(i)
    sum+=val
    if val>max:
        max=val
    if val<min:
        min=val
        
print(max)
print(min)
print(sum)
```
